###Folha

```python
import requests
from bs4 import BeautifulSoup
requisicao = requests.get('https://www1.folha.uol.com.br/tec/')

import pandas as pd
```

```python
#titulos1
```

```python
html = requisicao.content
sopa = BeautifulSoup(html)
```

```python
titulos = sopa.findAll('h2', {'class': 'c-headline__title'})
```

```python
titulos_raspados = []

for titulo in titulos[:5]:
  titulos_raspados.append(titulo.text.strip())
print(titulos_raspados)
```

```python
df_titulos = pd.DataFrame({'Titulo': titulos_raspados})
df_titulos
```

```python
#links1
```

```python
links_raspados = []

for div in sopa.find_all('div', {'class': 'c-headline__content'})[:5]:
    a_tag = div.find('a')  # pega a <a> dentro da div
    if a_tag and 'href' in a_tag.attrs:
        links_raspados.append(a_tag['href'])

print(links_raspados)
```

```python
df_links = pd.DataFrame({'Link': links_raspados})
df_links
```

```python
#df final 1
```

```python
#df 1
```

```python
df_final_folha = pd.concat([df_titulos, df_links], axis=1)
df_final_folha['Veículo'] = 'Folha'
df_final_folha
```

###TeleSíntese

```python
requisicao2 = requests.get('https://telesintese.com.br/plantao-de-noticias/')
```

```python
html2 = requisicao2.content
sopa2 = BeautifulSoup(html2)
```

```python
#titulos 2
```

```python
titulos2 = sopa2.findAll('h2', {'class': 'entry-title'})
```

```python
titulos_raspados2 = []

for titulo in titulos2[:5]:
  titulos_raspados2.append(titulo.text.strip())
print(titulos_raspados2)
```

```python
titulos_raspados2
```

```python
df_titulos2 = pd.DataFrame({'Titulo': titulos_raspados2})
df_titulos2
```

```python
#links 2
```

```python
links2 = sopa2.find_all('a', class_='ct-media-container')

links_raspados2 = []

for link in links2[:5]:
    href = link.get('href')
    if href:
        links_raspados2.append(href.strip())

print(links_raspados2)
```

```python
df_links2 = pd.DataFrame({'Link': links_raspados2})
df_links2
```

```python
#df 2
```

```python
df_final_telesintese = pd.concat([df_titulos2, df_links2], axis=1)
df_final_telesintese['Veículo'] = 'TeleSintese'
df_final_telesintese
```

###Teletime

```python
requisicao3 = requests.get('https://teletime.com.br/noticias/')
```

```python
html3 = requisicao3.content
sopa3 = BeautifulSoup(html3)
```

```python
#titulos 3
```

```python
titulos3 = sopa3.findAll('h3', {'class': 'entry-title td-module-title'})
```

```python
titulos_raspados3 = []

for titulo in titulos3[:5]:
  titulos_raspados3.append(titulo.text.strip())
print(titulos_raspados3)
```

```python
df_titulos3 = pd.DataFrame({'Titulo': titulos_raspados3})
df_titulos3
```

```python
#links3
```

```python
links3 = sopa3.find_all('h3', class_='entry-title td-module-title')

links_raspados3 = []

for link in links3[:5]:
    a_tag = link.find('a')
    if a_tag and a_tag.get('href'):
        href = a_tag.get('href').strip()
        links_raspados3.append(href)

print(links_raspados3)
```

```python
df_links3 = pd.DataFrame({'Link': links_raspados3})
df_links3
```

```python
#df 3
```

```python
df_final_teletime = pd.concat([df_titulos3, df_links3], axis=1)
df_final_teletime['Veículo'] = 'Teletime'
df_final_teletime
```

###DF final

```python
teste_df_final = pd.concat([df_final_folha, df_final_telesintese, df_final_teletime])
teste_df_final
```

```python
from datetime import datetime
from zoneinfo import ZoneInfo

agora = datetime.now(ZoneInfo("America/Sao_Paulo"))

data = agora.strftime("%d-%m-%Y")
hora = agora.strftime("%H-%M")

nome_arquivo = f"planilha noticias - data {data} - {hora}.csv"

teste_df_final.to_csv(
    nome_arquivo,
    index=False
)
```
